```python
from bs4 import BeautifulSoup
import re
import pandas as pd
import matplotlib.pyplot as plt
import requests
```

```python
request = requests.get('https://www.flipkart.com/search?q=laptops&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page=31')
soup = BeautifulSoup(request.content, 'html.parser')
scrap = soup.findAll('a', href=True, attrs={'class':'_31qSD5'})
```

```python
l=[]
base_url = "https://www.flipkart.com/search?q=laptops&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page="
for page in range(1,32):
    request = requests.get(base_url + str(page))
    soup = BeautifulSoup(request.content, 'html.parser')
    scrap = soup.findAll('a', href=True, attrs={'class':'_31qSD5'})
    for a in scrap:
        d = {}
        d['Name'] = a.find('div', {'class':'_3wU53n'}).text.split('-')[0]
        d['Price'] = a.find('div', {'class':'col col-5-12 _2o7WAb'}).find('div', {'class':'_1vC4OE _2rQ-NK'}).text
        d['Processor'] = a.find('div', {'class':"_3ULzGw"}).find('ul', {'class':'vFw0gD'}).findAll('li', {'class':'tVe95H'})[0].text
        d['RAM'] = a.find('div', {'class':"_3ULzGw"}).find('ul', {'class':'vFw0gD'}).findAll('li', {'class':'tVe95H'})[1].text
        try:
            d['Operating_system'] = a.find('div', {'class':"_3ULzGw"}).find('ul', {'class':'vFw0gD'}).findAll('li', {'class':'tVe95H'})[2].text
        except IndexError:
            pass
        try:
            d['Hard_disk'] = a.find('div', {'class':"_3ULzGw"}).find('ul', {'class':'vFw0gD'}).findAll('li', {'class':'tVe95H'})[3].text
        except IndexError:
            pass
        try:
            d['Dispay_inch'] = a.find('div', {'class':"_3ULzGw"}).find('ul', {'class':'vFw0gD'}).findAll('li', {'class':'tVe95H'})[4].text
        except IndexError:
            pass
        l.append(d)
```

```python
df = pd.DataFrame(l)
```
