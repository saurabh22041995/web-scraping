```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np

request = requests.get('http://www.investors-clinic.com/Greater%20Noida?page=6')

soup = BeautifulSoup(request.content, 'html.parser')

scrap = soup.find_all('div',{'class':'wht-cont'})
```

```python
l = []
base_url = "http://www.investors-clinic.com/Greater Noida?page="
for page in range(1,7):
    request = requests.get(base_url + str(page))
    soup = BeautifulSoup(request.content, 'html.parser')
    scrap = soup.find_all('div',{'class':'wht-cont'})
    for item in scrap:
        d = {}
        d['Society Name'] = item.find_all('a',{'class':'text-red'})[0].text
        d['Rate'] = item.find_all('p',{'class':'text-white'})[0].text.replace('\t','').split(',')[0].split('-')[1]
        d['Area'] = item.find_all('p',{'class':'text-white'})[0].text.replace('\t','').split(',')[1]
        d['Description'] = item.find_all('p',{'class':'text-white'})[0].text.replace('\t','').split(',')[2].split('\n')[0]
        d['Address'] = item.find_all('p',{'class':'text-white'})[0].text.replace('\t','').split(',')[2].split('\n')[1]
        d['City'] = item.find_all('p',{'class':'text-white'})[0].text.replace('\t','').split(',')[3].replace('\n', '')
        l.append(d)
```

```python
df = pd.DataFrame(l)
```
